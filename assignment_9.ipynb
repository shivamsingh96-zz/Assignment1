```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv')
```

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv')
```

```python
result = pd.concat([df1, df2])
```

```python
result.isnull().sum()
```

```python
result['Survived'].isnull()
```

```python
result['Survived'].dropna()
```

```python
result['Has_Cabin']  = result['Cabin'].notnull()
```

```python
result.head()
```

```python
# Split passenger titles in Name category and place in new 'TitleF' column
result['Title'] = result['Name'].map(lambda name: name.split(',')[1].split('.')[0].strip())
# Print unique passenger titles and their counts
result['Title'].value_counts() 
```

```python
# Create new 'TitleS' column to store shortened title from full 'TitleF' column
result['TitleS'] = result['Title']
# Replace adult female titles with 'Mrs'
result['TitleS'] = result['TitleS'].replace(['Lady', 'Mlle', 'Mme', 'Ms', 'the Countess'], 'Mrs')
# Replace adult male titles with 'Mr'
result['TitleS'] = result['TitleS'].replace(['Capt', 'Col', 'Don', 'Jonkheer', 'Major', 'Rev', 'Sir'], 'Mr')
# Replace dona titles with 'Special'
result['TitleS'] = result['TitleS'].replace(['Dona'], 'Special')
# Replace females with title of Dr to Mrs
result['TitleS'] = result['TitleS'].replace(result['TitleS'].loc[(result.Sex == 'female') &
 (result.Title == 'Dr')], 'Mrs')
# Replace males with title of Dr to Mr
result['TitleS'] = result['TitleS'].replace(result['TitleS'].loc[(result.Sex == 'male') &
 (result.Title == 'Dr')], 'Mr')
# Print unique passenger TitleS
result['TitleS'].value_counts()
```

```python
result
```

```python
# fill missing values with mean column values
result.fillna(result.mean(), inplace=True)
```

```python
result
```

```python
result.isnull().sum()
```

```python
result['CategoricalAge'] = pd.cut(result['Age'], 5)

print (result[['CategoricalAge', 'Fare']].groupby(['CategoricalAge'], as_index=False))
```

```python
result['CategoricalFare'] = pd.qcut(result['Fare'], 4)
print (result[['CategoricalFare', 'Age']].groupby(['CategoricalFare'], as_index=False))
```

```python
result
```

```python
irrelevant_features = ['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin','Age','Fare']
```

```python
result = result.drop(irrelevant_features, axis=1)
print (result.head())
```

```python
survived = df1[df1['Survived']==1]
not_survived = df1[df1['Survived']==0]
print ('Survived:\n', survived.head(), '\nNot Survived:\n', not_survived.head())
```

```python
survived = result[result['Survived']==1]
not_survived = result[result['Survived']==0]
print ('Survived:\n', survived.head(), '\nNot Survived:\n', not_survived.head())
```

```python
survived.head()
```

```python
print ('Survived: %i, (%f%%)' %(len(survived), float(len(survived)*100/len(result))))
print ('Not Survived: %i, (%f%%)' %(len(not_survived), float(len(not_survived)*100/len(result))))
print ('Not Survived: %i' %(len(result)))
```

```python
# Transform into binary variables
result_dum = pd.get_dummies(result, drop_first=True)
result_dum.head()
```

```python
x = result_dum.iloc[:, 1:].values
y = result_dum.iloc[:, 0:1].values
print (x.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2, random_state=0)
print (x_train.shape, y_train.shape)
```

```python
y_test.shape
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(x_test)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
classifier.score(x_test,y_test)
```
